##  Task- 3- Outlier Detection


```python
#  Identify outlier countries in South Asia based on their Score and GDP per Capita.
Q1_score = df_southasia['score'].quantile(0.25)
Q3_score = df_southasia['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

Q1_gdp = df_southasia['Log GDP per capita'].quantile(0.25)
Q3_gdp = df_southasia['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

# Identify outliers based on IQR
outliers_score_iqr = df_southasia[(df_southasia['score'] < Q1_score - 1.5 * IQR_score) | (df_southasia['score'] > Q3_score + 1.5 * IQR_score)]
outliers_gdp_iqr = df_southasia[(df_southasia['Log GDP per capita'] < Q1_gdp - 1.5 * IQR_gdp) | (df_southasia['Log GDP per capita'] > Q3_gdp + 1.5 * IQR_gdp)]

print("Outliers based on IQR for Score:")
print(outliers_score_iqr)

print("\nOutliers based on IQR for GDP per Capita:")
print(outliers_gdp_iqr)
```

```python
#Define outliers using the 1.5 × IQR rule.
import pandas as pd
import numpy as np

def detect_outliers(data):
  """Detects outliers in a given dataset using the 1.5 IQR rule.

  Args:
    data: A numpy array or pandas Series containing the data.

  Returns:
    A list of indices corresponding to the outliers.
  """

  q1 = np.percentile(data, 25)
  q3 = np.percentile(data, 75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5 * iqr
  upper_bound = q3 + 1.5 * iqr
  outliers = np.where((data < lower_bound) | (data > upper_bound))[0]
  return outliers

# Assuming you have a DataFrame named 'df_southasia'
outliers_score_indices = detect_outliers(df_southasia['score'])
outliers_gdp_indices = detect_outliers(df_southasia['Log GDP per capita'])

outliers_score = df_southasia.iloc[outliers_score_indices]
outliers_gdp = df_southasia.iloc[outliers_gdp_indices]

print("Outliers based on Score:")
print(outliers_score)

print("\nOutliers based on GDP per Capita:")
print(outliers_gdp)
```

```python
# Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
#in a different color
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'df_southasia'

# Identify outliers using the 1.5 IQR rule
outliers_score_indices = detect_outliers(df_southasia['score'])
outliers_gdp_indices = detect_outliers(df_southasia['Log GDP per capita'])

# Create a scatter plot
plt.figure(figsize=(10, 6))

# Plot all data points in one color
plt.scatter(df_southasia['Log GDP per capita'], df_southasia['score'], color='blue', label='Non-Outliers')

# Plot outliers in a different color
plt.scatter(df_southasia.iloc[outliers_score_indices]['Log GDP per capita'],
            df_southasia.iloc[outliers_score_indices]['score'],
            color='red', label='Outliers (score)')

plt.scatter(df_southasia.iloc[outliers_gdp_indices]['Log GDP per capita'],
            df_southasia.iloc[outliers_gdp_indices]['score'],
            color='red', label='Outliers (Log GDP per capita)')

plt.xlabel('Log GDP per capita')
plt.ylabel('score')
plt.title('Scatter Plot of score vs.Log GDP per capita with Outliers')
plt.legend()
plt.grid(True)
plt.show()
```

Cultural and Religious Factors:

Cultural values and religious beliefs can influence people's perceptions of happiness and well-being.
Countries with strong social support systems and a sense of community may have higher happiness levels, even if their economic indicators are relatively low.
Health and Education:

Access to quality healthcare and education can significantly impact people's well-being.
Countries with well-developed healthcare and education systems tend to have higher happiness levels.
Policy Implications:

To bridge the gap between economic growth and happiness, policymakers should consider the following:

Promoting equitable growth: Implement policies to reduce income inequality and ensure that the benefits of economic growth are shared by all.
Investing in social infrastructure: Improve access to healthcare, education, and social services.
Strengthening governance and institutions: Promote good governance, reduce corruption, and ensure the rule of law.
Protecting the environment: Implement sustainable development policies to protect the environment and mitigate climate change.
Promoting social cohesion and community engagement: Foster a sense of belonging and social connectedness.
